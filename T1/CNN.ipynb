```python
import os
import shutil
from sklearn.model_selection import train_test_split

# Function to split and copy dataset
def split_dataset(dataset_path, base_dir, train_ratio=0.7, valid_ratio=0.2, test_ratio=0.1):
    # Ensure base directories exist
    train_dir = os.path.join(base_dir, 'train')
    test_dir = os.path.join(base_dir, 'test')
    valid_dir = os.path.join(base_dir, 'validation')

    for directory in [train_dir, test_dir, valid_dir]:
        if not os.path.exists(directory):
            os.makedirs(directory)

    # Iterate over classes
    for class_folder in os.listdir(dataset_path):
        class_path = os.path.join(dataset_path, class_folder)

        if os.path.isdir(class_path):
            # Get all images in the class folder
            all_images = []
            for root, dirs, files in os.walk(class_path):
                for file in files:
                    if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
                        all_images.append(os.path.join(root, file))

            # Split into train, validation, and test
            train_images, temp_images = train_test_split(all_images, test_size=(1 - train_ratio), random_state=42)
            valid_images, test_images = train_test_split(temp_images, test_size=(test_ratio / (test_ratio + valid_ratio)), random_state=42)

            # Copy files to respective directories
            for image_set, target_dir in zip([train_images, valid_images, test_images], [train_dir, valid_dir, test_dir]):
                class_target_dir = os.path.join(target_dir, class_folder)
                if not os.path.exists(class_target_dir):
                    os.makedirs(class_target_dir)
                for image_path in image_set:
                    shutil.copy(image_path, os.path.join(class_target_dir, os.path.basename(image_path)))
                    print(f"Copied {image_path} to {class_target_dir}")

# Example usage
dataset_path = r"D:\P\AI Digital Assignment\Dataset"  # Replace with your dataset path
base_dir = r"D:\P\AI Digital Assignment\Dataset_final"  # Replace with your base directory
split_dataset(dataset_path, base_dir, train_ratio=0.7, valid_ratio=0.2, test_ratio=0.1)
```

```python
base_dir = r"D:\P\AI Digital Assignment\Dataset_final"
train_dir = os.path.join(base_dir, 'train')
test_dir = os.path.join(base_dir, 'test')

# Image preprocessing parameters
IMG_HEIGHT = 224
IMG_WIDTH = 224
BATCH_SIZE = 32
EPOCHS = 20

# Data generators with augmentation for training and validation sets
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # Split for validation
)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',  # Multi-class labels
    subset='training'
)

validation_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_HEIGHT, IMG_WIDTH),
    batch_size=BATCH_SIZE,
    class_mode='categorical',  # Multi-class labels
    subset='validation'
)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score, roc_auc_score
from sklearn.preprocessing import label_binarize
```

```python
y_true = validation_generator.classes  # True labels
y_pred = model.predict(validation_generator)
y_pred_classes = np.argmax(y_pred, axis=1)  # Predicted classes

# Calculate accuracy
accuracy = accuracy_score(y_true, y_pred_classes)
print(f"Validation Accuracy: {accuracy:.2f}")

# Calculate F1-score (weighted)
f1 = f1_score(y_true, y_pred_classes, average='weighted')
print(f"F1-Score (Weighted): {f1:.2f}")

# Confusion Matrix
conf_matrix = confusion_matrix(y_true, y_pred_classes)

# Plot heatmap of confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', 
            xticklabels=validation_generator.class_indices.keys(),
            yticklabels=validation_generator.class_indices.keys())
plt.title('Confusion Matrix Heatmap')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()

# AUC-ROC (One-vs-Rest for multi-class)
n_classes = len(validation_generator.class_indices)  # Number of classes
y_true_binarized = label_binarize(y_true, classes=range(n_classes))  # Binarize true labels

# Compute AUC for each class
auc_scores = []
for i in range(n_classes):
    auc = roc_auc_score(y_true_binarized[:, i-1], y_pred[:, i])
    auc_scores.append(auc)
    print(f"AUC-ROC for Class {i} ({list(validation_generator.class_indices.keys())[i]}): {auc:.2f}")

# Plot AUC-ROC curves for all classes
plt.figure(figsize=(10, 8))
for i in range(n_classes):
    from sklearn.metrics import roc_curve
    fpr, tpr, _ = roc_curve(y_true_binarized[:, i], y_pred[:, i])
    plt.plot(fpr, tpr, label=f'Class {i} ({list(validation_generator.class_indices.keys())[i]}): AUC = {auc_scores[i]:.2f}')

plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.title('AUC-ROC Curves (One-vs-Rest)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.show()
```
